```julia
using DataFrames, Gaston
```

```julia
function saveplot(str)
    Gaston.gnuplot_send("set terminal eps mono")
    s= "set output '"*str*"'"
    Gaston.gnuplot_send(s)
    Gaston.gnuplot_send("replot")
    Gaston.gnuplot_send("set terminal windows")
    Gaston.gnuplot_send("set output")
end
```

```julia
set(pointsize=2,grid="on")
```

# 非反転増幅回路

```julia
niac=readtable("Non-inverting amplifier circuit_1k.txt", separator='\t')
niacm=readtable("Non-inverting amplifier circuit_1M.txt", separator='\t')
niac[:A]=1e3*niac[:V_pp_out]./niac[:V_pp_in]
niacm[:A]=1e3*niacm[:V_pp_out]./niacm[:V_pp_in]

amp(x)=1+160.4/x
amp2(x)=1/(1/10^(15/20)+x/(x+160.4))

plot(Array(niac[:R_1]), Array(niac[:A]), plotstyle="points", legend="1[kHz]に対する測定値",xlabel="抵抗[kΩ]",ylabel="増幅率",xrange="[5:16]")
plot!(Array(niacm[:R_1]), Array(niacm[:A]), plotstyle="points",legend="1[MHz]に対する測定値",marker="x")

t = 5:0.01:16
plot!(t,amp.(t), legend="1[kHz]に対する理論値")
plot!(t,amp2.(t), legend="1[MHz]に対する理論値")
saveplot("niac_r.eps")
```

# 周波数特性
* 増幅率

```julia
fc=readtable("Frequency characteristic.txt", separator='\t')
fc[:A]=1e3*fc[:V_pp_out]./fc[:V_pp_in]

plot(Array(fc[:Frequency]), Array(20*log10.(fc[:A])),axis="semilogx",plotstyle="points",xlabel="周波数[Hz]",ylabel="増幅率[dB]",yrange="[0:30]",xrange="[50:2e6]")
saveplot("niac_f.eps")
```

* 位相

```julia
fc[:ph]=2*π*fc[:Frequency].*fc[:Delay]*1e-6
plot(Array(fc[:Frequency]), Array(fc[:ph]),plotstyle="points",axis="semilogx",xlabel="周波数[Hz]",ylabel="位相差[rad]",yrange="[-1.6:0.1]",xrange="[50:2e6]")
saveplot("niac_phi.eps")
```

# 移相回路

```julia
pts=100
R=15.905e3
C=9.86e-9
t=R*C
x=[10^(6*i/pts) for i in 0:pts]
psc=readtable("Phase-shift circuit.txt" ,separator='\t')
psc[:ph]=2*π*psc[:Frequency].*psc[:Delay]
psc[:A]=psc[:V_pp_out]./psc[:V_pp_in]

g(x)=t*2*π*x/(sqrt((1-t^2*(2*π*x)^2)^2+9t^2*(2*π*x)^2)) #amp

plot(x,g.(x),axis="semilogx",legend="理論値",xlabel="周波数[Hz]",ylabel="振幅比")
plot!(Array(psc[:Frequency]), Array(psc[:A]),plotstyle="points",legend="測定値")
saveplot("psc_amp.eps")
```

```julia
f(x)=atan((1-t^2*(2*π*x)^2)/(6*π*t*x)) #phase

plot(x,f.(x),axis="semilogx",legend="理論値",xlabel="周波数[Hz]",ylabel="位相差[rad]")
plot!(Array(psc[:Frequency]), Array(psc[:ph]),plotstyle="points",legend="測定値")
saveplot("psc_ph.eps")
```

# ウィーンブリッジ発振回路
* 発振周波数

```julia
osc=readtable("oscillator.txt" ,separator='\t')
h(a)=sqrt(6a-a^2-5)/4/t/π

x=collect(2.95:0.01:3.35)
plot(x,h.(x),xlabel="増幅率",ylabel="発振周波数[Hz]",legend="理論値",yrange="[970:1020]")
plot!(Array(osc[:A]), Array(osc[:f]),legend="測定値",plotstyle="points")
saveplot("osc.eps")
```

* 緩和時間

```julia
rt=readtable("Relaxation time.txt" ,separator='\t')
rt[:A]=1+12016./rt[:R_1]
plot(Array(rt[:A]), Array(rt[:t]),xlabel="周波数",ylabel="緩和時間[s]",plotstyle="points")
saveplot("rt.eps")
```

緩和時間は(A-3)/τに反比例するのだから、かければ、一定のはずだけど、、

```julia
rlx(a)=(3.007685-a)/t
rt[:pro]=convert(Array{Float64},(rt[:t].*rlx.(rt[:A])))

plot(Array(rt[:A]),Array(rt[:pro]),xlabel="増幅率",ylabel="積",plotstyle="points")
saveplot("rt_pro.eps")
```
